```python
# library import
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn

```

```python
# leitura dos dados

data = pd.read_csv("/kaggle/input/ahpcrit/CAHP2.csv")
```

```python
print(data)
```

```python
# análise de correlação

correlation = data.corr(method='pearson')
```

```python
# plot da matriz de correlação entre critérios
plot = sn.heatmap(correlation, annot = True, fmt=".1f", linewidths=.8)
plot
plt.savefig('pearson.png', dpi=600) 
```
